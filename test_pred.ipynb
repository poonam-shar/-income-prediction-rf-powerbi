```python
import pandas as pd
import numpy as np
import joblib

# Load the trained model
model = joblib.load("income_predictor_model.pkl")

# Load the test dataset
test_df = pd.read_csv("au_test.csv")

# Clean the test data
test_df = test_df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
test_df.replace('?', np.nan, inplace=True)

# Save original test data index for reference
test_index = test_df.index

# Predict using the trained model
predictions = model.predict(test_df)
predicted_probabilities = model.predict_proba(test_df)[:, 1]  # Probability of class 1 (>50K)

# Create a DataFrame with results
results_df = test_df.copy()
results_df['Predicted_Class'] = predictions
results_df['Probability_>50K'] = predicted_probabilities

# Map back prediction to label
results_df['Predicted_Label'] = results_df['Predicted_Class'].map({0: '<=50K', 1: '>50K'})

# Show first few rows
results_df[['Predicted_Class', 'Probability_>50K', 'Predicted_Label']].head()
```
